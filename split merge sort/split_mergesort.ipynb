```python
import pandas as pd

### use merge sort. two functions, one sorts and another splits

# split and merge
def split_merge(_array):
    if len(_array) < 2:
        return _array
    # find middle point by halving the length of _array, and return integer
    mid = round(len(_array)/2)
    # recursively call split_merge while invoking sort_array function with splitted _array 
    return sort_array(
        left_array = split_merge(_array[:mid]),
        right_array = split_merge(_array[mid:]) 
        )

# sort 
def sort_array(left_array, right_array):
    if len(left_array)==0:
        return right_array
    elif len(right_array)==0:
        return left_array
    
    # initialize result_array
    result_array = []

    # initialize left and right indices to 0
    left_index = 0
    right_index = 0

    # loop
    while len(result_array) < len(left_array) + len(right_array):
        # compare right and left elements
        # print(left_array[left_index], right_array[right_index])
        if left_array[left_index] <= right_array[right_index]:
            result_array.append(left_array[left_index])
            left_index+=1
        else: 
            result_array.append(right_array[right_index])
            right_index+=1
        # leftover items
        if len(left_array)==left_index:
            # append rest of right_array[right_index:] to result_array
            result_array+=right_array[right_index:]
            break
        if len(right_array)==right_index:
            # append rest of left_array[left_index:] to result_array
            result_array+=left_array[left_index:]
            break
    return result_array

# iterate through each species and columns
df = pd.read_csv('iris.csv')
result = pd.DataFrame()

for spc in df.species.unique():
    sample = df.loc[df['species']==spc, df.columns!='species']
    sorted_feature = {}
    for col in sample.columns:
        # apply split_merge function to series
        sorted_feature[(spc, col)] = split_merge(sample[col].tolist())
        
    # merge sorted series as dataframe
    class_result = pd.DataFrame(sorted_feature)
    # merge sorted species dataframes into one.
    result = pd.concat([result, class_result], axis=1)

result
```
